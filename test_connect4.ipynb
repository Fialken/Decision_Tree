```python
from DecisionTree import DTree
from DecisionTree import Evaluation as EVA

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("datasets/connect4.csv", header=None)
dfy = df[42]
df.drop(columns=42,inplace=True)


#dividir em sets de treino e teste
X_train, X_test, y_train, y_test = train_test_split(df, dfy, test_size=0.3,random_state=321)


arvore = DTree()
```

```python
arvore.create_DTree(X_train,y_train)
print(arvore)
```

```python
pred = arvore.predict(X_test)
EVA.accuracy(pred,y_test)
print()
EVA.confusion_matrix(pred,y_test)
```
